```python
# !pip install pymupdf
# !pip install arxiv

from langchain_community.document_loaders import TextLoader
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.document_loaders import WebBaseLoader
import bs4
from langchain_community.document_loaders import ArxivLoader
#import time
from langchain_community.document_loaders import WikipediaLoader


## Text Loader
loader = TextLoader('Gautam_buddha.txt')
loader
text_doc = loader.load()
text_doc

## PDF Loader
loader = PyPDFLoader('sample-1.pdf')
pdf_doc = loader.load()
pdf_doc

## Web Based Loader
web_loader = WebBaseLoader(web_paths=("https://en.wikipedia.org/wiki/Main_Page",),
        bs_kwargs=dict(parse_only=bs4.SoupStrainer(
            class_=("mw-jump-link","mw-page-container","mw-footer-container")   # "mw-jump-link","mw-page-container","mw-footer-container" - these are the css classess in wiki page
        ))
            )
web_loader.load()

## Arxiv
#time.sleep(3)
'''
docs = ArxivLoader(query="attention is all you need", load_max_docs=2).load()
len(docs)
print (docs[0])
'''

## Wiki loader
wiki_docs = WikipediaLoader(query="Generative AI", load_max_docs=2).load() 
len(wiki_docs)
print (wiki_docs[0])
```
