Sentence Transformers on Hugging Face

```python
# https://huggingface.co/models

import os
from dotenv import load_dotenv
from langchain_openai import OpenAIEmbeddings

load_dotenv()     # Loading env variables

os.environ['HUGGING_FACE_ACCESS_TOKEN']=os.getenv("HUGGING_FACE_ACCESS_TOKEN")


```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

text = "This is a text for Hugging Face Embedding Example."
query_result=embeddings.embed_query(text)
query_result
len(query_result)
```

```python
# With List of multiple documents
doc_result = embeddings.embed_documents([text,"Siddhartha Gautama, most commonly referred to as the Buddha was a wandering ascetic."])
doc_result[0]
```
