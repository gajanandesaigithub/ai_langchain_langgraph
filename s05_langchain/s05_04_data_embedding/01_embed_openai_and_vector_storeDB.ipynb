```python
import os
from dotenv import load_dotenv
from langchain_openai import OpenAIEmbeddings

load_dotenv()     # Loading env variables



os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings

text = "Siddhartha Gautama, most commonly referred to as the Buddha was a wandering ascetic."
query_result = embeddings.embed_query(text) # This will convert to vectors
query_result 

len(query_result)  # Check dimension of result = 3072 default for this model

# We can also set our own dimension
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
query_result = embeddings_1024.embed_query(text)
len(query_result)
```

```python
# We can convert text to vector for an entire document
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

## Text Loader
loader = TextLoader('Gautam_buddha.txt')
docs = loader.load()
docs

# Split it into data chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20)
text = text_splitter.split_documents(docs)
text
```
