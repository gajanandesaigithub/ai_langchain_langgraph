```python
class Error(Exception):
    pass

class dobException(Error):
    pass

year = int(input("Enter Birth year"))
age = 2026-year

try: 
    if age >=20 and age <=30:
        print("You are eligible for exam.")
    else:
        raise dobException

except dobException:
    print("Sorry, your age condition is not fulfilling the exam eligibility criteria. Age should be between 20 to 30 years.")
```
