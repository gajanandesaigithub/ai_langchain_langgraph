```python
import logging

## Configure the basic logging settings
logging.basicConfig(level=logging.DEBUG)



## Configuring logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True                          #In Python logging, logging.basicConfig() only works once per process. force = True will 1) Remove existing handlers, 2) Reconfigure logging using your settings
)

logging.debug("This is a debug message")
logging.info("This is a info message")
logging.warning("This is a warning message")
logging.error("This is a error message")
logging.critical("This is a critical message")

#-------
# Usually we create logging.ipynb in logs folder and add abve configuration in it.
# See the logs folder where we have created logger.py and test.py. When we run test.py, it will create app.log file inside log folder and write logs in it.
#-------
```
