```python
##------------------------------------
## Arithmatic Operations
##------------------------------------

a = 10
b = 4

addition_result = a + b
subtraction_result = a - b
multiplication_result = a * b
division_result = a/b
floor_division_result = a//b
modulus_result = a%b
exponential_result = a**b       # 'a' to the of power 'b'

print("Addition of a & b is ", addition_result)
print("Subtraction of a & b is ", subtraction_result)
print("Multiplication of a & b is ", multiplication_result)
print("Division of a & b is ", division_result)
print("Floor Division of a & b is ", floor_division_result) 
print("Modulus of a & b is ", modulus_result)
print("Exponential of a & b is ", exponential_result)


##------------------------------------
## Comparision Operators
##------------------------------------
print ("\n------------------------------------\n**Comparision Operators**\n------------------------------------")
num1 = 40
num2 = 20

print("\nnum1 = ", num1, ", num2 = ", num2)
print("num1 = num 2?", (num1==num2))
print("num1 >= num2?", (num1 >= num2))
print("num1 < num2?", (num1 < num2))


##------------------------------------
## Logical Operators : and  or not
##------------------------------------

print ("\n------------------------------------\n**Logical Operators**\n------------------------------------")
rainyDay = True
sunnyDay = False
print("Value of rainyDay = ", rainyDay)
print("Value of sunnyDay = ", sunnyDay)
print("Rainy Day AND Sunny Day:", rainyDay and sunnyDay)
print("Rainy Day OR Sunny Day:", rainyDay or sunnyDay)
print("NOT Rainy Day: ", not rainyDay)
```
