```python
### Abstraction in Python
# Hiding complex implementation features and showing only the necessary features.

from abc import ABC, abstractmethod
import math

## Abstract base class
class Shape(ABC):           ## Define asbtract class by inheriting ABC
    def color(self):
        pass

    @abstractmethod
    def area (self):
        pass

class Square(Shape):        ##
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2

class Circle (Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius

c = Circle(3)
s = Square(4)
print ("Areas of circle with radius 3 is : ",c.area())
print ("Areas of Square with side 4 is : ",s.area())
```
