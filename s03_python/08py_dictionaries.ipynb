```python
###-------------------------------------
###     Definition of Dictionaries in Python
###-------------------------------------
dict_1 = {}        ## Empty Dictionary
dict_2 = dict()    ## Empty Dictionary

print("Type of dict_1",type(dict_1))
print("Type of dict_2", type(dict_2))

dict_buddy = {"name":"Aniket", "age":25, "hobby":"Music"}
print("dict_buddy:",dict_buddy)
print(type(dict_buddy))

###-------------------------------------
###     Accessing elements of Python Dictionaries
###-------------------------------------
print("Buddy name:",dict_buddy['name'])
print("Buddy age:",dict_buddy['age'])
print("Buddy hobby:",dict_buddy['hobby'])

###-------------------------------------
###     Accessing elements of Python Dictionaries using get method
###-------------------------------------
print("Buddy name:",dict_buddy.get('name'))
print("Buddy age:",dict_buddy.get('age'))
print("Buddy hobby:",dict_buddy.get('hobby'))
print("Buddy hobby:",dict_buddy.get('lst_name'))                    ## This prints 'None' because this key does not exist in dictionary
print("Buddy hobby:",dict_buddy.get('lst_name',"Not present"))      ## We can provide default value like "Not Present" for keys that does not exist in dictionary

###     Modifying and Adding elements of Python Dictionaries using get method
dict_buddy["age"]=26                    ## Update existing value
dict_buddy["address"]="Mumbai"          ## Add new key-value
print(dict_buddy)                       ## Delete key-value

del dict_buddy['address']
print(dict_buddy)

###-------------------------------------
###     Methods of Dictionaries
###-------------------------------------

buddy_keys   = dict_buddy.keys()                    ## Get all keys
buddy_values = dict_buddy.values()                  ## Get all values
all_items    = dict_buddy.items()                   ## Get all key-value pairs
print("All keys of buddy dict:", buddy_keys)
print("All values of buddy dict:", buddy_values)
print("All key value pairs of buddy dict:", all_items)

dict_buddy_copy = dict_buddy.copy()                 ## Make shallow copy of dictionary. This makes two separate copies
dict_buddy_copy["name"]="Aniket111"                 ## In shallow copy, changing items of one does not change on others
print("dict_buddy: ", dict_buddy)
print("dict_buddy_copy: ", dict_buddy_copy)
```
