```python
###-------------------------------------
###     Definition of Dictionaries in Python
###-------------------------------------
dict_1 = {}        ## Empty Dictionary
dict_2 = dict()    ## Empty Dictionary

print("Type of dict_1",type(dict_1))
print("Type of dict_2", type(dict_2))

dict_buddy = {"name":"Aniket", "age":25, "hobby":"Music"}
print("dict_buddy:",dict_buddy)
print(type(dict_buddy))

###-------------------------------------
###     Accessing elements of Python Dictionaries
###-------------------------------------
print("Buddy name:",dict_buddy['name'])
print("Buddy age:",dict_buddy['age'])
print("Buddy hobby:",dict_buddy['hobby'])

###-------------------------------------
###     Accessing elements of Python Dictionaries using get method
###-------------------------------------
print("Buddy name:",dict_buddy.get('name'))
print("Buddy age:",dict_buddy.get('age'))
print("Buddy hobby:",dict_buddy.get('hobby'))
print("Buddy hobby:",dict_buddy.get('lst_name'))                    ## This prints 'None' because this key does not exist in dictionary
print("Buddy hobby:",dict_buddy.get('lst_name',"Not present"))      ## We can provide default value liek "Not Present" for keys that does not exist in dictionary

```
