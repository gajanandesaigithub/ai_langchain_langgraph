```python
### Iterators are advanced Python concepts that allow for effient looping and memory management.
# Iterators provide a way to access elements of collection sequencially without exposing the underlying structure.

numbers = [1,2,3,4,5,6]
print(type(numbers))
print(numbers)
for i in numbers:
    print(i)

## Iterator
iterator = iter(numbers)

print(type(iterator))

```

```python
## iterate through all the elements
try:
    print(next(iterator))                          ## Ieteratos use lazy loading. on next() only next element from the list is loaded in memory. ## Once all elements are fetched one by one then it will reach end of iterator and next() will gove error "StopIteration".
except StopIteration:
    print("There are no elements in iterator.")                                    
```

```python
## String iteration
my_string = "Greetings!"
string_iterator = iter(my_string)

print(next(string_iterator))   # Output G
print(next(string_iterator))   # Output r
```
