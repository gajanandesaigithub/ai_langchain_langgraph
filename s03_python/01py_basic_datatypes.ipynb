```python
## Integer Examples

age = 40
print("\nAge=",age)
print("Datatype of variable 'age' = ", type(age))

## Floating point datatype
height = 5.61
print("\nHeight=",height)
print("Datatype of variable 'height' = ", type(height))

## String  datatype
name = 'Gajani'
print("\nName=",name)
print("Datatype of variable 'name' = ", type(name))

## Boolean  datatype
enrolledCourse = True
print("\nHas Enrolled Course? =",enrolledCourse)
print("Datatype of variable 'enrolledCourse' = ", type(enrolledCourse))

## Default value retyrned by bool() is False
flag = bool()
print("\nflag = ", flag)

a = 10
b = 10
print("(a == b)? ", (a==b))
```
