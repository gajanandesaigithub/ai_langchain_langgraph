```python
## Standard libraries in Python
import array
import math
import random
import os           ## For file and directory access
import shutil       ## high level operations on files and collections of files
import json         ## data serialization
import csv

arrNumEven = array.array('i',[2,4,6,8,10])
print(arrNumEven)

print(math.sqrt(5))

print(random.randint(1,10))
print(random.randint(1,10))
print(random.choice(['Red','Blue','Green','Yellow','Orange','Purple','White']))
print(os.getcwd())        ## Get current working directory

##shutil.copyfile('source.txt', 'destination.txt')

## Convert dictionary into json
data = {'name':'Aniket', 'Age':25}
json_str = json.dumps(data)                
print(json_str)
print(type(json_str))

## Convert json back to dictionary
parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))


## CSV
## create new file and open it in write mode
with open('example.csv',mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Aniket',25])

with open('example.csv',mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
