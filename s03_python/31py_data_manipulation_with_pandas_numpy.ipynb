```python
import pandas as pd

df = pd.read_csv('customers-100.csv')
df.head(5)      # fetch first five rows
df.tail(5)      # fetch last five rows
df.describe()   # get stats
df.dtypes       # get datatypes of fields
df.isnull()     # returns entriew table with true / false instead of values. True=null value
df.isnull().any(axis = 1)   # If true if any value in a row is null else false. axis=1 ==> operate row wise.
df.isnull().any(axis=0)     # If true if any value in a column is null else false. axis=0 ==> operate column wise.
df_removed_blank = df[~df.isnull().any(axis=1)]   # Remove rows with null values
df.isnull().sum()       # Give count of null values in each column
df_filled = df.fillna(0)            # fill null values with zero.

# filling missing values with mean of the column
df['Buying_fill_NA'] = df['Monthly Buying'].fillna(df['Monthly Buying'].mean)

# Renaming column
df = df.rename(columns = {'Monthly Buying':'Buying'})

```
