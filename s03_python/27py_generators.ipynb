```python
## Geretators are simpler way to create iterators. They use the yeild keyowrd to produce a series of values 
## lazily, which means they generate values on the fly and do not store then in memory.
# 

#---------------------------------------------------------------
# Example 1 Suppose we want squares of a series of numbers
#---------------------------------------------------------------
def square(n):
    for i in range(n):
        return (i**2)
## But this method will quit the method with square of only first number. But we want square of all multiple numbers.
print("Calling square",square(3))

## Solution: We use generator with yeild keyword which will generate the square, 
# then ut will store it in local variable or yeild the result to the caller and return back to the for loop in the method
def square_new(n):
    for i in range(n):
        yield (i**2)

for i in square_new(3):
    print(i)

## If we want to assign the result  to varibale, we can do that e.g.
a = square_new(4)
print(type(a))
print (a) 
print(next(a))
print(next(a))
print(next(a))
print(next(a))
#print(next(a))          # End of iterator.. error - StopIteration

#---------------------------------------------------------------
# Example 2
#---------------------------------------------------------------
def my_generator():
    yield 1
    yield 2
    yield 3
    
gen = my_generator()
print(next(gen))
print(next(gen))
print(next(gen))
#print(next(gen))   # End of iterator.. error - StopIteration

## Using for loop to display values
def my_generator1():
    yield 1
    yield 2
    yield 3
    
gen = my_generator1()
for val in gen:
    print(f"Using for loop:{val}")


#---------------------------------------------------------------
# Example 3: Practical example. Reding large file, without loading the content of entire file
#---------------------------------------------------------------
def read_large_file(file_path):
    with open(file_path) as file:
        for line in file:
            yield line
```
