```python
##--------------------------------
## Inheritance
##--------------------------------
class Furniture:
    def __init__(self, furType, furMaterial, furHeight):
        self.furType = furType
        self.furMaterial = furMaterial
        self.furHeight = furHeight

    def furLife(self):
        if (self.furMaterial.lower())=="wooden":
            print(f"Life of {self.furMaterial} {self.furType} is 50 years.")
        else:
            print(f"Life of {self.furMaterial} {self.furType} is less than 50 years.")

## Inheriting Furniture class into Chair class
class Chair(Furniture):
    def __init__(self, furType, furMaterial, furHeight, isFoldable):    ## Constructor of child class (derived class)
        super().__init__(furType, furMaterial, furHeight)               ## Calling constructor of parent class (base class)
        self.isFoldable = isFoldable

    def getFoldableInfo(self):
        print(f"Is Chair foldable? {self.isFoldable}")

## Instanciating
chair1 = Chair("Dining Chair", "Wooden", "2.5 feet", "Yes")
chair1.getFoldableInfo()
chair1.furLife()


##----------------------------------------------------------------
## Multiple Inheritance: Inheriting more than one base classes
##----------------------------------------------------------------
```
