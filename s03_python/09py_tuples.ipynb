```python
'''
Tuples are immutable. Lists are mutable.
Tuples cannot grow/shrink. Lists can.
Tuples are hashable. Lists are not.
Tuples can be used as keys. Lists cannot be.
Tuples are used for fixed structure. Lists are used as sequence.

With a tuple, Python promises:
    - values wonâ€™t change
    - hash can be trusted
    - safe to use as a key
    - safe to share across logic

Use List when it is like "This is a collection of things that may change"
Use Tuple when it is like "This is a single thing made of parts"

'''

### --------------------------
### Defining Tuples
### --------------------------
'''
## Using ()
tpl_1 = ()
print(type(tpl_1))
print(tpl_1)

## Using tuple() class
tpl_2 = tuple();            
print(type(tpl_2))
print(tpl_2)

## Using list
lst = [1,2,3,4,5]
print(type(lst))
print(lst)
tpl_3 = tuple(lst)
print(type(tpl_3))
print(tpl_3)

## Similarly we can also convert list using tuples
lst = list((1,2,3,4,5))
print(type(lst))
print(lst)

### Tuples can contain elements of mixed data types
tpl_4 = (23, "Vishnu", 67.5, True)
print(type(tpl_4))
print(tpl_4)
'''

'''
### --------------------------
### Accessing elements in Tuples.
### --------------------------

## we access tuple elements with [], similar to Lists. Even slicing of tuples is similar to that of Lists.
tpl_5 = (100, 321, 453, 567)
print(tpl_5[0])
print(tpl_5[-1])        ## Access last element with index -1
print(tpl_5[0:2])       ## Get elements till index < 2
print(tpl_5[::])        ## Get all elements
print(tpl_5[::-1])      ## Get elements is reverse order
'''
### --------------------------
### Operations on  Tuples.
### --------------------------

##  Concatinate two or more tuples with +
combined_tpl = tpl_4 + tpl_5
print(type(combined_tpl))
print(combined_tpl)

## Replicate elements in Tuples with *
print(tpl_5 * 2)


### --------------------------
### Immutability of Tuples - We cannot modify elements in Tuple
### --------------------------

## The below statement will give TypeError when we try to change element value in tuple
tpl_6 = (1,2,3, 45,45)
##tpl_6[0] = 50

### --------------------------
### Methods of Tuples
### --------------------------
tpl_6.count (45)    ## get count of 45 in the tuple
tpl_6.index(45)     ## get index of the element 45

### --------------------------
### Packing and Unpacking of Tuples
### --------------------------

## Packing
## When we define elements comma separated without brackets, they are by default convereted (packed) into Tuple

tpl_7 = 123, "Vishnu", 342.8, False
print(type(tpl_7))
print(tpl_7)

## UnPacking
a,b,c,d=tpl_7
print("unpacked value into a =",a)
print("unpacked value into b =",b)
print("unpacked value into c =",c)
print("unpacked value into d =",d)

## Unpacking with help of *
colors = ("red", "green", "blue", "yellow", "orange", "violet", "white")
var1, *var2, var3 = colors   ## Start varibale will be assigned all rest of the elements
print (var1)
print (var2)
print (var3)




```
