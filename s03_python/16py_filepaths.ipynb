```python
import os  
## Create new directory
'''
new_dir = "new_folder"
os.mkdir(new_dir)
'''

## Listing all dir and files from a directory
items = os.listdir('.')    ## '.' is current location
print (items)


## Joining paths

## 1. Get relative path
dir_name = "folder"
file_name = "myFile.txt"
full_path = os.path.join(dir_name, file_name)
print(full_path)

## 2. get full path
full_path = os.path.join(os.getcwd(), dir_name, file_name)    ## get full path by os.getcwd()
print(full_path)

## Check if file already exists
path = "myFile.txt"
if os.path.exists(path):
    print(f"The path {path} already exists.")
else:
    print(f"The path {path} does not exist.")

## Check if the path is a directory or a file
path = "myFile.txt"
if os.path.isfile(path):
    print(f"The path {path} is a file.")
elif os.path.isdir(path):
    print(f"The path {path} is a directory.")
else:
    print(f"The path {path} is niether a file nor a directory.")

## Getting an absolute path
relative_path = "myFile.txt"
absolute_path = os.path.abspath(path)
print(absolute_path)


```
