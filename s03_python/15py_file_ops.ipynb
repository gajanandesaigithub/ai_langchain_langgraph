```python
###---------------------------
### File operations in Python
###---------------------------

### ---------- READ WHOLE FILE

'''
## Example: Create a text file manually financials.txt with below content
    A. Annual Financial Statement (AFS)
    B. Demands for Grants (DG)
    C. Finance Bill
'''
with open('financials.txt','r') as file:
    content = file.read()
    print(content)

### -------- READ file line by line
with open('financials.txt','r') as file:
    for line in file:
        print(line.strip())                 ## by default it will add new line at the end. .strip() will remove new line


### ---------- WRITING IN FILE
'''
with open('financials.txt','w') as file:            ## This will open file to overwrite the existing content
    file.write("E. Expenditure Budget\n")           
    
'''

with open('financials.txt','a') as file:            ## opening file in append mode.
    file.write("E. Expenditure Budget. (This is new line appended by opening file in append mode.)\n")           ## This will overwrite the existing content
   
```
