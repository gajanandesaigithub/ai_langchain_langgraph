```python
###---------------------------
### File operations in Python
###---------------------------

### ---------- READ WHOLE FILE

'''
## Example: Create a text file manually financials.txt with below content
    A. Annual Financial Statement (AFS)
    B. Demands for Grants (DG)
    C. Finance Bill
'''
with open('financials.txt','r') as file:
    content = file.read()
    print(content)

### -------- READ file line by line
with open('financials.txt','r') as file:
    for line in file:
        print(line.strip())                 ## by default it will add new line at the end. .strip() will remove new line


### ---------- WRITING IN FILE
'''
with open('financials.txt','w') as file:            ## This will open file to overwrite the existing content
    file.write("E. Expenditure Budget\n")           
    
'''

with open('financials.txt','a') as file:            ## opening file in append mode.
    file.write("E. Expenditure Budget. (This is new line appended by opening file in append mode.)\n")           ## This will overwrite the existing content


## Writing alist of files
lines = ['Its January month.\n', 'The weather is cool.\n', 'World is beautiful.\n']
with open('financials.txt','a') as file:
    file.writelines(lines)
'''
with open('financials.txt','r') as file:
    for line in file:
        print(line.strip())   
'''

## Binary files
## Binary files are .bin files. Binary files are updated in bytes.
## To write in binary files, they are opened in 'wb' or 'rb' i.e. Write Byte or Read Byte mode.
data = b'\x00\x01\x02\x03\x04'
with open ('binExample.bin','wb') as file:
    file.write(data)
with open('binExample.bin','rb') as file:
   content = file.read()
   print(content)


## Reading from a file and writing into another file.
with open('financials.txt','r') as source_file:
    content = source_file.read()
with open('destination.txt','w') as destination_file:
    destination_file.write(content)


## Example: Read the text file count the number of lines, words and characters.
def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines) 
        return line_count, word_count, char_count

file_path='financials.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines = {lines}, Words = {words},Characters = {characters}')
```
