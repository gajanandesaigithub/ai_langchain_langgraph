```python
## Encapsulation is a concept of wrapping data, variables and methods together as a single unit. 
# It restricts direct access to the some of the object components which is a mean of 
# preventing an accidentanl interference and misuse of data.


### Encapsulation

## Access modifiers: Public, private & protected
class Car:
    def __init__(self, name, engineType):
        self.name = name                    ## public variable
        self.engineType = engineType        ## public variable 

car = Car("SCross", "diesel")
print(car.name)
dir(car)                                    ## dir gives list of public methods and variables of the object


## Private variables
class Car1:
    def __init__(self, name, engineType):
        self.__name = name                    ## private variable... name starts with __ i.e. double unsderscore
        self.__engineType = engineType        ## private variable... name starts with __ i.e. double unsderscore

car1 = Car1("Mahindra XUV500", "petrol")
## print(car1.name)   we will get error on this statement or even on print(car1.__name) because now name is private variable.
dir(car1)               ## now dir() will list name and engineType as _Car1__engineType and _Car1__name
print(car1._Car1__name) ## This will return name... but this is not recommended

#### Even derived calss will not be able to access private variables of parent class.

## Protected variables: Not accesible to outside of class but accessible to derived class
class Car2:
    def __init__(self, name, engineType):
        self._name = name                    ## Protected variable... name starts with _ i.e. single unsderscore
        self._engineType = engineType        ## Protected variable... name starts with _ i.e. single unsderscore

car2 = Car2("Harrier.ev", "electrical")
dir(car2)               ## now dir() will list name and engineType as _Car1_engineType and _Car1_name
```
