```python
# Lambda functions in Python are defined by the keyword "lambda:" and can have any number of arguments but only one expression
# Syntaax
# lambda arguments: expression

# lambda function examples
multiplication = lambda x, y: x*y
print(type(multiplication))             # type is <class 'function'>
print(  multiplication(23, 10)  )       # Calling lambda function... just mention the the variable name and pass arguments in round brackets.

### maps() in Python
numbers = [1,2,3,4,5]

map(lambda x:x**2, numbers)                     # Read this as for each element in 'numbers' execute lambda, i.e. calculate square

squares = list(map(lambda x:x**2, numbers))     # We convert this to list and print to see the output
print(squares)

##---------------------------------
## Map with multiple iterables
##---------------------------------
numbers1 = [23, 34, 45, 50]
numbers2 = [ 3,  4,  5, 50,43]
diff_nums = map(lambda x,y:x-y, numbers1, numbers2)     ## THis will subtract each num of numbers2 from the respective number of numbers1, if there is corresponding element present in both

print(list(diff_nums))

## map() to convert string to integer
str_numbers = ['1', '2', '3', '4']
int_numbers = list(map(int, str_numbers))
print(int_numbers)

colors = ['red', 'blue', 'green']
colors_in_caps = list(map(str.upper, colors))
print(colors_in_caps)

##Applying map() to dictionary
def get_name (person):
    return person ['name']

people=[
    {'name':'Aniket', 'age':25},
    {'name':'Kishor','age':40},
    {'name':'Sandeep','age':42}
]

names = list(map(get_name, people))
print(list(map(get_name, people)))
```
