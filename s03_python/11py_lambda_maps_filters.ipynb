```python
# Lambda functions in Python are defined by the keyword "lambda:" and can have any number of arguments but only one expression
# Syntaax
# lambda arguments: expression

# lambda function examples
multiplication = lambda x, y: x*y
print(type(multiplication))             # type is <class 'function'>
print(  multiplication(23, 10)  )       # Calling lambda function... just mention the the variable name and pass arguments in round brackets.

### maps() in Python
numbers = [1,2,3,4,5]

map(lambda x:x**2, numbers)                     # Read this as for each element in 'numbers' execute lambda, i.e. calculate square

squares = list(map(lambda x:x**2, numbers))     # We convert this to list and print to see the output
print(squares)
```
