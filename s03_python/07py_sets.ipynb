```python
###------------------------------------
###      SETS Definition
###------------------------------------
print("\n\n\n###------------------------------------\n###      SETS Definition\n###------------------------------------")

'''
my_set1 = {1,1,2,3}
print("my_set1:",my_set1)                 ## This will ignore duplicates and print only unique values
print("Type of my_set1:",type(my_set1))

my_set2 = set({1,2,3,4,5,6})  ## Defining with help of set class and set (i.e. with curly braces)
print("my_set2:",my_set2)
print("Type of my_set2:",type(my_set2))


my_set3 = set([1,2,3,4,5,6])  ## Defining with help of set class and list (i.e. with square braces)
print("my_set4:",my_set3)
print("Type of my_set3:",type(my_set3))
'''



###------------------------------------
###      Basic Operations on Sets
###------------------------------------
print("\n\n\n###------------------------------------\n###      Basic Operations on Sets\n###------------------------------------")
print("\n## Adding, Removing and Discarding elements of sets")

'''
## Adding Elements
my_set4 = set([1,2,3,4,5,6])
print("my_set4:",my_set4)
my_set4.add(7)
print("my_set4 after addition of elements:",my_set4)

## Removing Elements
my_set4.remove(7)
print("my_set4 after addition of elements:",my_set4)

## Removing Elements
    ## ** If you try to remove element which is not present in the set, it will give error.
    ## HENCE WE USE DISCARD
my_set4.discard(10)
'''

'''
## Popping an element from the set
    ## ** Popping will remove the element arbitrarily from the set. Arbitrarily does not means randomly but it means implementation-defined, not user-defined.

    ## Let us consider we add elements in this order: 10, 0, 9, 1, 8, 2, 7, 3, 6, 4, 5
my_set5 = {10}
my_set5.add(0)
my_set5.add(9)
my_set5.add(1)
my_set5.add(8)
my_set5.add(2)
my_set5.add(7)
my_set5.add(3)
my_set5.add(6)
my_set5.add(4)
my_set5.add(5)

print(my_set5)

    ## Now we pop one by one and see which element is popped out
popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)

popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)

popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)

popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)

popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)

popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)

popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)

popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)

popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)

popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)

popped_element = my_set5.pop()
print("Popped element:", popped_element)
print("Set after pop:",my_set5)
'''


###------------------------------------
###      Set Membership Test
###------------------------------------
print("\n\n\n###------------------------------------\n###      Set Membership Test\n###------------------------------------")

my_set6 = {2,3,5,8}
print ("2 is member of my_set6: ", 2 in my_set6)
print ("10 is member of my_set6: ", 10 in my_set6)


###------------------------------------
###      Mathematical Operations: Union, Intersection, Intersection_update, Difference, Difference_update, Symmetric_Difference
###------------------------------------
print("\n\n\n###------------------------------------\n###      Mathematical Operations on Sets \n###------------------------------------")

set_A = {1,2,3,4,5,6}
set_B = {4,5,6,7,8,9}

print("Set A: ", set_A)
print("Set B:", set_B)
print("Union of Set A and Set B: ", set_A.union(set_B))                                 ## Prints unique elements of A & B together
print("Difference of Set A and Set B: ", set_A.difference(set_B))                       ## Prints elements of A after removing common elements of B
print("Difference of Set B and Set A: ", set_B.difference(set_A))                       ## Prints elements of B after removing common elements of A
print("Intersection of Set A and Set B: ", set_A.intersection(set_B))                   ## Prints commmon elements
print("Symmetric Difference of Set A and Set B: ", set_A.symmetric_difference(set_B))   ## Removes common elements and prints rest unique elements

## difference_update, intersection_update and symmetric_difference_update will assign the differece to the set on which these operations are performed.

###------------------------------------
###      Set Methods
###------------------------------------
print("\n\n\n###------------------------------------\n###      Set Methods\n###------------------------------------")
'''
.issubset()
.issuperset()
etc...
'''

###------------------------------------
###      Set Assignment: Counting unique words in text
###------------------------------------
print("\n\n\n###------------------------------------\n###      Set Assignment: Counting unique words in text \n###------------------------------------")
text = "Learning is boring process. Embrace this boredom. If you can sit this boredom, you will outgrow 90 percent of people."
words = text.split()            ## Get the list of words
unique_words = set(words)       ## Convert list to sets to remove duplicates
print(unique_words)
print("Count of words in original text:",len(words))
print("Count of words unique words:",len(unique_words))
```
