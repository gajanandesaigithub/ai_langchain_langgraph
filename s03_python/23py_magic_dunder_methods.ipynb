```python
'''
Magic methods in Python, also known as dunder methods (double underscore methods) are special methods that start and end with double underscore.
These methods enable to define the behavior of objects for built-in operations, such as airthmatic operations, comparisons etc.

__init__    : Initializes a new instance of a class.
__str__     : Returns a string representation of an object.
__repr__    : Returns an official string representation of an object.
__len__     : Returns the length of an object.
__getitem__ : Gets an item from a container.
__setitem__ : Sets an item in a container.

'''
##We can get list of magic methods by using dir() method
class Shape:
    pass

shape = Shape()
dir(shape)


class Vehicle():        
    def __init__(self, engine_type):
        self.engine_type = engine_type

car = Vehicle("electric")
print(car.engine_type)

print(car)                      ## this will print something like. <__main__.Vehicle object at 0x000002786D0C42F0>

##we can override magic fuctions
class Vehicle1():        
    def __init__(self, engine_type):
        self.engine_type = engine_type
    
    def __str__(self):
        return f"The vehicle is {self.engine_type} type."

    ## similarly we can override any dunder methods in our own way.

car1 = Vehicle1("electric")
print(car1.engine_type)

print(car1)                      ## Now we have overridden __str__ mthod so it will print customized message instaead of <__main__.Vehicle object at 0x000002786D0C42F0>

```
