```python
## LISTS ARE MUTABLE, MEANING WE CAN MODIFY LISTS

## Defining list

myList=[]               ## Empty List
print(type(myList))     ## Print type of list
print(myList)           ## print elements of list

## Lists can contain elements of different data types
lstPersonalDetails = ["Sandeep", 55, 165.5, True]
print(lstPersonalDetails)

## Accessing list elements
lstColors = ["Red", "Blue", "Green", "White", "Purple", "Orange", "Yellow"]
print(lstColors,"\n\n\n\n ")


'''
print("\n-----------------------------------------\n*** Printing elements in normal sequence***\n-----------------------------------------")
print(lstColors[0])
print(lstColors[1])
print(lstColors[2])
print(lstColors[3])
print(lstColors[4])
print(lstColors[5])
print(lstColors[6])
'''

'''
print("\n-----------------------------------------\n*** Printing elements in reverse sequence***\n-----------------------------------------")

## index -1 means last index.

print(lstColors[-1])
print(lstColors[-2])
print(lstColors[-3])
print(lstColors[-4])
print(lstColors[-5])
print(lstColors[-6])
print(lstColors[-7])
'''

###-------------------------------------------
###      ACCESSING LIST ELEMENTS
###-------------------------------------------
'''
print("\n-----------------------------------------\n*** Printing part of list sequence***\n-----------------------------------------")
print(lstColors[3:])        ## This prints elemenets with index 3 onwards
print(lstColors[-1:])       ## This prints elements with index -1 (stat one) onwards, meaning only last element
print(lstColors[3:6])       ## This prints elemenets with index 3, 4 and 5. And NOT till 6
print(lstColors[4:-1])      ## index -1 means last index. This prints elemenets with index 4 onwards till second last element
'''

###-------------------------------------------
###      MODIFYING LIST ELEMENTS
###-------------------------------------------
'''
print("\n-----------------------------------------\n*** MODIFYING LIST ELEMENTS ***\n-----------------------------------------")
lstColors[4]="Magenta"
print(lstColors)

lstColors[1:] = "Golden"     ## NOTE: The result will be ['Red', 'G', 'o', 'l', 'd', 'e', 'n']
print(lstColors)
'''

###-------------------------------------------
###      APPEND ELEMENTS TO LIST AT THE END
###-------------------------------------------
'''
print("\n-----------------------------------------\n*** APPEND ELEMENTS TO LIST***\n-----------------------------------------")
lstColors = ["Red", "Blue", "Green", "White", "Purple", "Orange", "Yellow"]
lstColors.append("Golden")      ## This will append the "Golden" color at the end
print(lstColors)
'''
###-------------------------------------------
###      INSERTING ELEMENTS TO LIST
###-------------------------------------------
'''
print("\n-----------------------------------------\n*** INSERT ELEMENTS IN THE MIDDLE OF LIST ***\n-----------------------------------------")
lstColors.insert(1,"Brown")      ## This will insert the "Brown" color at second place
print(lstColors)
'''
###-------------------------------------------
###      REMOVING ELEMENTS FROM THE LIST
###-------------------------------------------
'''
print("\n-----------------------------------------\n*** REMOVE FIRST OCCURANCE OF ELEMENT FROM THE LIST ***\n-----------------------------------------")
lstColors.remove("Brown")      ## This will remove the "Brown" color from the list
print(lstColors)
'''

###-------------------------------------------
###      POPPING ELEMENTS FROM THE LIST (REMOVE AND GET THE REMOVED ELEMENT IN VARIABLE)
###-------------------------------------------
'''
print("\n-----------------------------------------\n*** REMOVE AND RETURN LAST ELEMENT ***\n-----------------------------------------")
print("List before popped color:",lstColors)
popped_color = lstColors.pop()
print("Popped color:",popped_color)
print("List after popped color:",lstColors)
'''

###-------------------------------------------
###      COUNTING ELEMENTS IN THE LIST
###-------------------------------------------
'''
print("Count of Blue color entris in list=",lstColors.count("Blue"))
'''

###-------------------------------------------
###      SORTING ELEMENTS IN THE LIST (Ascending sort)  
###-------------------------------------------
'''
lstColors.sort()
print("Sorted color list: ",lstColors)
'''

###-------------------------------------------
###      RECERSING ELEMENTS IN THE LIST (Desending sort)  
###-------------------------------------------
'''
lstColors.reverse()
print("Descending sorted list: ", lstColors)
'''

###-------------------------------------------
###      CLEARING ALL ELEMENTS IN THE LIST  
###-------------------------------------------
'''
lstColors.clear()
print("Printing list after clearing off all elements in the list: ", lstColors)
'''


###-------------------------------------------
###      SLICING LIST  
###-------------------------------------------
'''
print("\n-----------------------------------------\n*** SLICING LIST ***\n-----------------------------------------")
print("Complete list:",lstColors)
print("lstColors[3:5]: ",lstColors[3:5])    ## Prints elements with index 3,4
print("lstColors[3:]",lstColors[3:])        ## Prints elements from 3 till end
print("lstColors[:4]",lstColors[:4])        ## Prints elements from index 0 to 3
print("lstColors[:-1]",lstColors[:-1])      ## Prints elements from index 0 to index max-1
print("lstColors[::2]",lstColors[::2])      ## Prints elements from 0 to last index with step of 2
print("lstColors[::-2]",lstColors[::-2])      ## Prints elements in reverse order from max index to first index with step of 2
'''
###-------------------------------------------
###      ITERATING THROUGH LIST  
###-------------------------------------------

print("\n-----------------------------------------\n*** Simple Iterating List using for loop ***\n-----------------------------------------")
for color in lstColors:
    print(color)

print("\n-----------------------------------------\n*** Iterating List with index using for loop ***\n-----------------------------------------")
for index,color in enumerate(lstColors):    ## Here we need to use enumerate() to track the index while iterating
    print (index,color)


###-------------------------------------------
###      LIST COMPREHENSION  
###      Syntax is like [ expression for item in iterable ] 
###-------------------------------------------


[x**2 for x in range(10)]               ## Here in single step we create list of squares of 0 to 9
print (  [x**2 for x in range(10)] )    ## Here in
## Syntax is like [ expression for item in iterable ] 


###-------------------------------------------
###      LIST COMPREHENSION with condition 
###      Syntax is like [ expression for item in iterable condition] 
###-------------------------------------------
sqr_of_even_nums = [x**2 for x in range (10) if (x%2==0)]
print(sqr_of_even_nums)


###-------------------------------------------
###      NESTES LIST COMPREHENSION
###      Syntax is like [ expression for item1 in iterable1 for item2 in iterable2] 
###-------------------------------------------
## Example: Here if we want to pair items is 2 listes, we can do it as below
list1 = [1,2,3,4]
list2 = ['a','b','c','d']
pair = [ [i,j] for i in list1 for j in list2]
print (pair)


###-------------------------------------------
###      LIST COMPREHENSION WITH FUNCTION CALLS
###-------------------------------------------
words = ["Today", "is", "a", "lazy", "Saturday", "for", "learning"]
lengths = [len(word) for word in words]
print("Lenghts of words in the list words are: ", lengths)
```
