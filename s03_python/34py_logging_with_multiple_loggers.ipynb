```python
import logging

## Create alogger for module 1
logger1 = logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

## Create alogger for module 2
logger2 = logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

## Configuring logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True                          #In Python logging, logging.basicConfig() only works once per process. force = True will 1) Remove existing handlers, 2) Reconfigure logging using your settings
)

# Now we can log messages with different loggers
logger1.debug("This DEBUG message is logged with logger 1")
logger2.warning("This WARNING message is logged with logger 2")
```
