#### 2. Model with Optional Fields
Add optional fields using Python's Optional type.
Note: Pydantic validates daty even for optional fields when values are provided.

```python
from typing import Optional
class Employee(BaseModel):
    id: int
    name: str
    department: str
    salary: Optional [float] = None # Optional with default value
    is_active: Optional [float] = True # Optional with default value True

emp1 = Employee(id=1, name = "Tom", department="Sales")
print(emp1)
print(type(emp1))

emp2 = Employee(id=1, name = "Tom", department="Sales", salary=230000, is_active=False)
print(emp2)
print(type(emp2))
```

```python
from pydantic import BaseModel
from typing import List

class Classroom(BaseModel):
    room_mumber: str
    students: List[str] # List of strings
    capacity: int

class1 = Classroom(
    room_mumber="A201",
    students=["Krushna", "Gorakha", "Arvind"],
    capacity=25
)
print(class1)
```

#### 3. Model with Nested Models
Create complex structures with nested models


```python
from pydantic import BaseModel

class Address(BaseModel):
    street: str
    city: str
    zip_code: int

class Customer(BaseModel):
    customer_id: int
    name: str
    address: Address    #Nested Model

# Create customer with nested model Address
customer = Customer(
    customer_id=1,
    name="Seema",
    address={"street": "LBS Road", "city": "Mumbai", "zip_code": "400003"}
)
print(customer)  # Here for zip_code we provided numner in str format, pydantic converts that to int


```

#### Pydantic Fields: Customization and Constraints
The Field function in Pydantic enhances model fields beyond basic type hints by allowing you to specify validation rules, default values, aliases, and more.

```python
from pydantic import BaseModel, Field
class Item(BaseModel):
    name: str = Field(min_length=2, max_length=50) # On top of basic validation of type, we are also addition min and max length
    price: float = Field(gt=0, le=5000)
    quantity: int = Field(gt=0)

# Valid instance
item = Item(name = "Flute", price = 2546.00, quantity=2)
print(item)
```

```python
from pydantic import BaseModel, Field

class User(BaseModel):
    username: str = Field(..., description="Unique username for the user")
    age: int = Field(default=18, description="User's age, defaults to 18 yrs")
    email: str = Field(default_factory=lambda: "user@example.com", description="Default email address")

#Examples
user1 = User(username="alice")
print (user1) # username='alice' age=18 email='user@example.com'

user2 = User(username="Arun", age=37, email="arun@xyzdomain.com")
print(user2) # username='Arun' age=37 email='arun@xyzdomain.com'
```
